### Arguments par défaut

````python
def volume_cone(height, radius=2.4):
    pi = 3.14
    return pi * radius ** 2 * height / 3
````

````python
# Case1: radius is omitted in volume_cone function call.
v = volume_cone(height=9) # height is passed in by name
print('The volume of a cone is: {}' .format(v))
````

````python
# Case2: the default value of radius=2.4 is overwritten.
v = volume_cone(9, 3) # height and radius are passed in by position
print('The volume of a cone is: {}' .format(v))
````

### Arguments multiples `(*args)` et arguments pour les mots `(**kwargs)`

Note : Utilisés pour rendre la fonction flexible.

````python
def compute_sum(a, b):
    return a + b
print(compute_sum(4, 5))
````

````python
def compute_sum(*args):
    print('Type of iterable object args {}.' .format(type(args)))
    count = 0
    for a in args:
        count += a
    return count
````

````python
result = compute_sum(0.5, 2.25, 7)
print('Result: ', result)
````

**Obs:** L'argument `args` n'est qu'un nom, l'élément clé est l'opérateur de décompression `*`.

La formule suivante calcule un score _Z score_ et mesure la distance entre une seule valeur de données brutes et la moyenne de la population.

\begin{equation*}
z = \frac{X - \mu }{\sigma }
\end{equation*}

où:
* **_X_** est une valeur unique de données brutes
* `mu` est la moyenne de la population
* `sigma` est l'écart-type de la population

Pour trouver l'écart-type, l'équation ci-dessous reste à notre disposition:

\begin{equation*}
\sigma = \sqrt{\frac{\sum \left | X - \mu \right |^{2}}{N}}
\end{equation*}

Où **_N_** est le nombre de points de données dans la population.
